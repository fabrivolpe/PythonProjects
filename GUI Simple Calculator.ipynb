```python
from tkinter import *
from tkinter import ttk

root = Tk()
root.title('Calculator')

s = ttk.Style()
s.theme_use('clam')


#frame
calculator = ttk.Frame(root)
calculator.grid(column = 0, row = 0, sticky = 'nsew')


#a variable that changes every time you click the button
curnum = ''

#function that changes the number on entry
def typeChar(num):
    
    global curnum
    
    curnum = curnum + str(num)
    
    equation.set(curnum)

    
#function to solve the equation
def solveProblem():
    
    try:
    
        global curnum

        solution = str(eval(curnum))

        equation.set(solution)

        curnum = solution
        
    except:
        
        equation.set('error')
        
        curnum = ''

#function to clear all        
def clear():
    
    global curnum
    
    curnum = ''
    
    equation.set(curnum)
    
#function to turn number to percentage
def turnPerc():
    
    global curnum
    
    solution = str(eval(curnum + '/100'))
    
    equation.set(solution)
    
    curnum = ''
    
#function to delete last character
def delete():
    
    global curnum
    
    s = list(curnum)
    
    s.pop()
    
    curnum = ''.join(s)
    
    equation.set(curnum)

#textvariable for the entry
equation = StringVar()

#entry
entry = ttk.Entry(calculator, textvariable = equation)
entry.grid(column = 0, row = 1, sticky = 'ew', columnspan = 4)

#button0
bzero = ttk.Button(calculator, text = '0', command = lambda: typeChar(0))
bzero.grid(column = 1, row = 5, sticky = 'ew')

#button1
bone = ttk.Button(calculator, text = '1', command = lambda: typeChar(1))
bone.grid(column = 0, row = 4, sticky = 'ew')

#button2
btwo = ttk.Button(calculator, text = '2', command = lambda: typeChar(2))
btwo.grid(column = 1, row = 4, sticky = 'ew')

#button3
bthree = ttk.Button(calculator, text = '3', command = lambda: typeChar(3))
bthree.grid(column = 2, row = 4, sticky = 'ew')

#button4
bfour = ttk.Button(calculator, text = '4', command = lambda: typeChar(4))
bfour.grid(column = 0, row = 3, sticky = 'ew')

#button5
bfive = ttk.Button(calculator, text = '5', command = lambda: typeChar(5))
bfive.grid(column = 1, row = 3, sticky = 'ew')

#button6
bsix = ttk.Button(calculator, text = '6', command = lambda: typeChar(6))
bsix.grid(column = 2, row = 3, sticky = 'ew')

#button7
bseven = ttk.Button(calculator, text = '7', command = lambda: typeChar(7))
bseven.grid(column = 0, row = 2, sticky = 'ew')

#button8
beight = ttk.Button(calculator, text = '8', command = lambda: typeChar(8))
beight.grid(column = 1, row = 2, sticky = 'ew')

#button9
bnine = ttk.Button(calculator, text = '9', command = lambda: typeChar(9))
bnine.grid(column = 2, row = 2, sticky = 'ew')

#buttondecimalpoint
bpoint = ttk.Button(calculator, text = '.', command = lambda: typeChar('.'))
bpoint.grid(column = 0, row = 5, sticky = 'ew')

#plussign
bplus = ttk.Button(calculator, text = '+', command = lambda: typeChar('+'))
bplus.grid(column = 3, row = 2, sticky = 'ew')

#minussign
bminus = ttk.Button(calculator, text = '-', command = lambda: typeChar('-'))
bminus.grid(column = 3, row = 3, sticky = 'ew')

#multiplicationsign
btimes = ttk.Button(calculator, text = '*', command = lambda: typeChar('*'))
btimes.grid(column = 3, row = 4, sticky = 'ew')

#divisionsign
bdivided = ttk.Button(calculator, text = '/', command = lambda: typeChar('/'))
bdivided.grid(column = 3, row = 5, sticky = 'ew')

#buttonpower
bpower = ttk.Button(calculator, text = '^', command = lambda: typeChar('**'))
bpower.grid(column = 1, row = 6, sticky = 'ew')

#equalsign
bequals = ttk.Button(calculator, text = '=', command = solveProblem)
bequals.grid(column = 3, row = 6, sticky = 'ew')

#buttonclear
bclear = ttk.Button(calculator, text = 'clear', command = clear)
bclear.grid(column = 2, row = 5, sticky = 'ew')

#buttonpercentage
bperc = ttk.Button(calculator, text = '%', command = turnPerc)
bperc.grid(column = 0, row = 6, sticky = 'ew')

#buttondelete
bdel = ttk.Button(calculator, text = 'del', command = delete)
bdel.grid(column = 2, row = 6, sticky = 'ew')

#buttonsquareroot
sqrt = ttk.Button(calculator, text = 'âˆš', command = lambda: typeChar('**0.5'))
sqrt.grid(column = 0, row = 7, sticky = 'ew')


root.mainloop()
```
